```python
import numpy as np
import matplotlib.pyplot as plt

#Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#Define points and their classes
x_points = np.array([-2, -1, 1, 2])
y_points = np.array([0, 0, 1, 1]) # 0 is the negative class, 1 is the positive class

#Define weights w and bias b
parameters = [
    (1, 0),
    (-1, 0),
    (0.5, 0),
    (5, 0)
    ]

#Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Sigmoid Function with Different Parameters', fontsize=16)

for ax, (w, b) in zip(axs.ravel(), parameters):
    #Calculate the sigmoid function for each point
    x_range = np.linspace(-5, 5, 100) # selecting whole range of x values
    z = w * x_range + b # finding z values for the sigmoid function
    y = sigmoid(z)

    #Plot the sigmoid function
    ax.plot(x_range, y, label=f'w={w}, b={b}') # the blue line, sigmoid curve
    ax.scatter(x_points, sigmoid(w * x_points + b), color='red', label='Data Points') # red dots, data points
    ax.scatter(x_points, y_points, color='green', label='True Labels') # green dots, true labels
    ax.axhline(0.5, color='gray', linestyle='--')
    ax.axvline(0, color='gray', linestyle='--')
    ax.set_title(f'Sigmoid Function (w={w}, b={b})')
    ax.set_xlabel('x')
    ax.set_ylabel('Sigmoid(x)')
    ax.legend()
    ax.grid()
```
