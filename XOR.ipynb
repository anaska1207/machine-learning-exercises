```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Original Data
X = np.array([[1,1], [-1,-1], [1,-1], [-1,1]])
y = np.array([1,1,-1,-1])

#Plotting 2D Data
plt.figure(figsize=(8,4))
for i, label in enumerate(y):
    plt.plot(X[i,0],X[i,1], 'ro' if label==1 else 'bo')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('2D Data')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go

#Applying Transformation 2D -> 3D : x1, x2, x1*x2
X_transformed = np.hstack((X, (X[:,0]*X[:,1]).reshape(-1,1)))

#Plotting 3D Data
fig = go.Figure()

#Adding points to the plot
colors = ['red' if label == 1 else 'blue' for label in y]
fig.add_trace(go.Scatter3d(x=X_transformed[:,0], y=X_transformed[:,1], z=X_transformed[:,2],
                           mode='markers', marker=dict(size=5, color=colors)))

#Defining the surface
xx, yy = np.meshgrid(np.linspace(-2, 2, 50), np.linspace(-2, 2, 50))
zz = 0 * xx
fig.add_trace(go.Surface(x=xx, y=yy, z=zz, opacity=0.5, colorscale='Greens'))

#Setting the layout of the plot
fig.update_layout(scene=dict(xaxis_title='x1', yaxis_title='x2', zaxis_title='x3=x1*x2'),
                  title='3D Data with Transformation',
                  margin=dict(l=0, r=0, b=0, t=30))

fig.show()
```
